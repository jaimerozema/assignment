# Introduction

The paper by [Acemoglu et al. (2001)](https://www.aeaweb.org/articles?id=10.1257/aer.91.5.1369) has been one of the most influential studies on economic development to date. The authors revolutionized macroeconomic studies by being able to go beyond merely stating correlations. By introducing an instrumental variable approach, they were able to infer the effect of institutions on present-day economic performance. They found that in areas in which Europeans faced high mortality rates, the Europeans were more likely to set up extractive (bad) institutions centuries ago as they did not settle there. These institutions, in turn, seem to have had large effects on income per capita.

## Instumental variable

To test for the effect of instutions on long-run economic performance, and to avoid selection bias, we use settler mortality as an instrumental variable to isolate part of the variation that is delivered randomly.

### Assumptions
There are several assumptions that have to be fulfilled. Firstly, settler mortality has to be randomly assigned in order for the IV to isolate the variation that is random (independence assumption). Secondly, the settler mortality has to be a sufficiently strong instrument, which means that the portion of random variation in settler mortality should be large enough. This will be tested in the descriptive statistics below via means of a regression. Lastly, settler mortality may not affect present-day economic performance directly, only via the treatment (exclusion restriction).
We can only test the second assumption through statistical analysis, but we cannot test the independence and exclusion restriction with our data. However, regarding the exclusion restriction, it seems very unlikely that settler mortality of the first European settlers (centuries ago) would have a direct impact on GDP per capita in 1995.

## Correlation quality of institutions and GDP
Below we test how the quality of institutions in a country, approximated by taking the average protection against expropriation risk, correlates with the log of GDP per capita in 1995 (PPP). We see in the graph that there exists a clear positive correlation between the two variables. This means that when a country has a high average level of protection against expropriation risk, it is likely that this country has a high GDP per capita in 1995 (PPP). We can also observe this positive correlation in the naive regression we run below the graph. In this correlation we see that the coefficient of 0.5158 has both a large value and statistical significance. However, we must refrain from making claims regarding causation and the direction of causation, i.e. we do not know whether higher GDP per capita leads to better protection against expropriation risk or vice versa. We will disentangle this in the next steps by introducing an instrumental variable. 

In the formula underneath we can see what equation will python actually use when we run the command; a simple regression that will not inform us on the actual treatment effect as previously said. Note that this formula is slithly different from the one used in the paper since we worked with a smaller dataset and so we could not include the same vector of covariates that [Acemoglu et al. (2001)](https://www.aeaweb.org/articles?id=10.1257/aer.91.5.1369) used. 

Naive Regression: 
\begin{equation}
logpgp95_{i} = \alpha + \beta*avexpr_{i} + \epsilon_{i}
\end{equation}

where:
 - logpgp95 = log of PPP GDP per capita in 1995 (source: World Bank)
 - alpha and beta = coefficients
 - avexpr = average protection against expropriation risk
 - epsilon = error term
 - i = country i

```python
# correlation log GDP 1995 and av. protection against expropriation risk
x = df['avexpr']
y = df['logpgp95']

plt.xlabel('Average protection against expropriation risk')
plt.ylabel('log GDP 1995')

fit = np.polyfit(x,y,1)
fit_fn = np.poly1d(fit) 
plt.plot(x, y, 'bo', x, fit_fn(x), '--k')
plt.show()
```

```python
#naive regression

import statsmodels.formula.api as smf    # for the ols and robust ols model
import statsmodels.api as sm
formula = "logpgp95 ~ avexpr"
results = smf.ols(formula,df).fit()
print(results.summary())
```

## Correlation quality of institutions and settler mortality
Below we test the viability of our instrumental variable: settler mortality. Firstly, we test how the variable average protection against expropriation risk correlates with settler mortality. In the graph below we see that there exists a negative correlation between the two variables. This means that when a country has a high average level of protection against expropriation risk, it is likely that this country had a low level of settler mortality. We can also observe this negative correlation when we regress the log of per capita GDP in 1995 (PPP) on the log of settler mortality (reduced form; below the graph). In this case, we observe that the coefficient of -0.5483 is statistically significant at all confidence levels.

In the formula underneath we can see what equation will python actually use when we run the command. Also here we did not include the vector of covariates for the same problem as before.

Reduced Form (Intention to treat): 
\begin{equation}
logpgp95_{i} = \alpha + \beta*logem4_{i} + \epsilon_{i}
\end{equation}

where:
 - logpgp95 = log of PPP GDP per capita in 1995 (source: World Bank)
 - alpha and beta = coefficients
 - logem4 = log of settler mortality
 - epsilon = error term
 - i = country i


```python
# IV: correlation av. protection against expropriation risk and settler mortality
x = df['logem4']
y = df['avexpr']

plt.xlabel('Log settler mortality')
plt.ylabel('Average protection against expropriation risk')

fit = np.polyfit(x,y,1)
fit_fn = np.poly1d(fit) 
plt.plot(x, y, 'yo', x, fit_fn(x), '--k')
plt.show()
```

```python
# Reduced form

formula = "logpgp95 ~ logem4"
results = smf.ols(formula,df).fit()
print(results.summary())
```

# Results

Since we know that our instrumental variable (settler mortality) is strong enough, we proceed by producing the results from our first stage: regress the average protection against expropriation risk against the log of settler mortality. From this, we save the resulting fitted values of the average protection against expropriation risk in a new variable. Moreover, since the coeffcient of logem4 (-0.6314) is statistically significant at all confidence levels, we can conclude that the instrumental variable (settler mortality) is sufficiently strong.

This new variable, with the fitted values, will in turn be used for the second stage: regressing the log of per capita GDP in 1995 (PPP) against the fitted values of the average protection against expropriation risk. We see here that the coefficient is 0.8684 and that it both large and statistically significant at all confidence levels.

Again, in the formulas underneath we will see what equation will python actually use when we run the commands. Also here we did not include the vector of covariates for the same problem as before.

First Stage Regression: 

\begin{equation}
avexpr_{i} = \alpha + \beta*logem4_{i} + \epsilon_{i}
\end{equation}

Second Stage Regression (Local Average Treatment Effect):

\begin{equation}
logpgp95_{i} = \alpha + \beta*\overline{avexpr_{i}} + \epsilon_{i}
\end{equation}

where:
 - logpgp95 = log of PPP GDP per capita in 1995 (source: World Bank)
 - avexpr =  average protection against expropriation risk (with overline the fitted values from the first stage)
 - alpha and beta = coefficients
 - logem4 = log of settler mortality
 - epsilon = error term
 - i = country i

```python
#First stage: regress avexpr logem4

formula = "avexpr ~ logem4"
results = smf.ols(formula,df).fit()
print(results.summary())
df['avexpr_hat'] = results.fittedvalues #saving fitted values for second stage
```

```python
#Second stage: regress logpgp95 avexpr_hat (predicted values for avexpr)
formula = "logpgp95 ~ avexpr_hat"
results = smf.ols(formula,df).fit()
print(results.summary())
```

## Robustness: omitted-variable bias
Below we test whether including the latitude of the capital into the regression has an impact on the robustness of our results. We see here, by running both the first and second stage again with the latitude of the capital included in the regression, that lat_abst reduces the absolute value of the coefficient for settler mortality in the first stage. This means that the strength of the instrumental variable is overestimated. However, as we can see below in the results for the second stage, the new estimated coefficient (1.03) is still statistically significant after including lat_abst.

Underneath we can see the last two formulas, these are not used by [Acemoglu et al. (2001)](https://www.aeaweb.org/articles?id=10.1257/aer.91.5.1369). We used them to add to the statistical anlysis a simple robustness check. 

First Stage Regression: 

\begin{equation}
avexpr_{i} = \alpha + \beta*logem4_{i} + \gamma*lat_abst + \epsilon_{i}
\end{equation}

Second Stage Regression (Local Average Treatment Effect):

\begin{equation}
logpgp95_{i} = \alpha + \beta*\overline{avexpr_{i}} + \gamma*lat_abst + \epsilon_{i}
\end{equation}

where:
 - logpgp95 = log of PPP GDP per capita in 1995 (source: World Bank)
 - lat_abst = latitude of capital
 - avexpr =  average protection against expropriation risk (with overline the fitted values from the first stage)
 - alpha and beta = coefficients
 - logem4 = log of settler mortality
 - epsilon = error term
 - i = country i

```python
#omitted variable

#first stage
formula = "avexpr ~ logem4 + lat_abst"
results = smf.ols(formula,df).fit()
#print(results.summary())
df['avexpr_hat2'] = results.fittedvalues #saving fitted values for second stage

#second stage
formula = "logpgp95 ~ avexpr_hat2 + lat_abst"
results = smf.ols(formula,df).fit()
print(results.summary())
```
